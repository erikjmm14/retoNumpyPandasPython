a. Extraer información del archivo.

```python
import pandas as pd
vacunas = pd.read_csv('/content/sample_data/country_vaccinations.csv')
```

b. Mostrar la estructura y tipos de datos de cada columna para identificar qué operaciones puedes realizar con cada una de ellas, asegurándote que las columnas con fechas sean del tipo **datetime64**.

```python
vacunas.head()
```

```python
vacunas.info()
```

```python
vacunas['date'] = pd.to_datetime(vacunas['date'])
vacunas.info()
```

c. Determinar la cantidad de vacunas aplicadas de cada compañía (con base a como lo reporta cada país en la columna **vaccines**).

```python
vacunas[['total_vaccinations', 'vaccines']].groupby('vaccines').sum()
```

d. Obtener la cantidad de vacunas aplicadas en todo el mundo.

```python
print('Cantidad de vacunas aplicadas en todo el mundo: ', vacunas['total_vaccinations'].sum())
```

e. Calcular el promedio de vacunas aplicadas por país.

```python
vacunas[['total_vaccinations', 'country']].groupby('country').mean()
```

f. Determinar la cantidad de vacunas aplicadas el día 29/01/21 en todo el mundo.

```python
dia_especial = vacunas.loc[vacunas['date'] == '2021-01-29', ['total_vaccinations']]
print('Cantidad de vacunas aplicadas el día 29/01/2021: ', dia_especial['total_vaccinations'].sum())
```

g. Crear un dataframe nuevo denominado **conDiferencias** que contenga los datos originales y una columna derivada (diferencias) con las diferencias de aplicación entre las columnas **daily_vaccinations** y **daily_vaccinations_raw**.

```python
conDiferencias = vacunas.copy()
conDiferencias['diferencias'] = conDiferencias['daily_vaccinations'] - conDiferencias['daily_vaccinations_raw']
conDiferencias.head()
```

h. Obtener el periodo de tiempo entre el registro con fecha más reciente y el registro con fecha más antigua.

```python
print('Periodo de tiempo entre el registro con fecha más reciente y más antigua: ', vacunas['date'].max() - vacunas['date'].min())
```

i. Crear un dataframe nuevo denominado **conCantidad** que contenga los datos originales y una columna derivada (canVac) con la cantidad de vacunas utilizadas cada día (usar la columna vaccines y separar por el caracter /).

```python
conCantidad = vacunas.copy()
conCantidad['canVac'] = conCantidad['vaccines'].str.split('/').str.len()
conCantidad.head()
```

j. Crear un dataframe nuevo denominado **antes20** con todos los registros que se hayan realizado antes del 20 de diciembre del 2020.

```python
antes20 = vacunas.loc[vacunas['date'] < '2020-12-20']
antes20.head()
```

k. Obtener un dataframe denominado **pfizer** con todos los registros donde se haya utilizado la vacuna Pfizer.

```python
pfizer = vacunas[vacunas['vaccines'] == 'Pfizer/BioNTech']
pfizer.head()
```

l. Almacenar los dataframes generados (conDiferencias, conCantidad, antes20 y pfizer) en un archivo de Excel denominado resultadosReto.xlsx, donde cada dataframe ocupe una hoja diferente.

```python
from pandas.io.excel import ExcelWriter
with pd.ExcelWriter('/content/reto_pandas.xlsx') as writer:
  conDiferencias.to_excel(writer, sheet_name='conDiferencias')
  conCantidad.to_excel(writer, sheet_name='conCantidad')
  antes20.to_excel(writer, sheet_name='antes20')
  pfizer.to_excel(writer, sheet_name='pfizer')
```
